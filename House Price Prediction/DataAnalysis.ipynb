### Missing Values Representation

```python
# Find the percentage of missing values in each feature

feature_null = [features for features in dataset.columns if dataset[features].isnull().sum() > 1]

# Print the feature name and % of missing values
print("Percentage of Missing Values:")

for feature in feature_null:
    print(f" {feature}: {np.round(dataset[feature].isnull().mean(), 4) * 100}%")
```

```python
for feature in feature_null:
    data = dataset.copy()

    #Variable that indicates 1 if observations is missing and 0 if it is available
    data[feature] = np.where(data[feature].isnull(), 1, 0)

    #Calculate the mean sale price when is missing
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

- The relation between the missing values and dependent variable in clearly visible.
- We need to replace these nan values with somevalues

### Numerical Variables

```python
# List of numerical Variables
numerical_features = [features for features in dataset.columns if dataset[features].dtypes != 'O']

print(f"Number of Numerical Features: {len(numerical_features)}")

data[numerical_features].head()
```

### Temporal Variables [Eg: Datetime Varibales]

- We have 4 year variables
- We can extract information from datatime variables such as no of years, months and days.

- Example: Age of the house

```python
#List of features containing years

year_features = [feature for feature in numerical_features if "Yr" in feature  or "Year" in feature]

year_features
```

```python
for feature in year_features:
    print(feature, dataset[feature].unique())
    print()
```

```python
# Check if there is a relation between year the house is sold and SalePrice

dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title("House Price vs Year Sold")
```

```python
# Compare the difference between all year features and Sale Price

for feature in year_features:
    if feature != 'YrSold':
        data = dataset.copy()
        data[feature] = data['YrSold'] - data[feature]

        plt.scatter(x = data[feature], y = data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel("Sale Price")
        plt.title(f"{feature} Vs Sale Price")

        plt.show()
```

### Discrete Variables

```python
discrete_features = [feature for feature in numerical_features if len(dataset[feature].unique()) < 25 and feature not in year_features + ['Id']]

print(f"Number of Discrete Features: {len(discrete_features)}")
```

```python
discrete_features
```

```python
dataset[discrete_features].head()
```

```python
# Relationship between Discrete Features and Sale Price 

for feature in discrete_features:
    data = dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.title(f"{feature} Vs Sale Price")

    plt.show()
```

### Continous Variables

```python
continous_features = [feature for feature in numerical_features if feature not in discrete_features + year_features + ['Id']]

print(f"Number of Continous Features: {len(continous_features)}")
```

```python
dataset[continous_features].head()
```

```python
for feature in continous_features:
    data = dataset.copy()
    data[feature].hist(bins = 25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    
    plt.show()
```

```python
# Applying Logarithmic Normal Distribution

for feature in continous_features:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['SalesPrice'] = np.log(data['SalePrice'])

        plt.scatter(data[feature], data["SalePrice"])
        plt.xlabel(feature)
        plt.ylabel("Sale Price")
        plt.title(f"{feature} Vs Sale Price")

        plt.show()
```

### Outliers

```python
for feature in continous_features:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        
        data.boxplot(column= feature)
        plt.title(feature)

        plt.show()
```
